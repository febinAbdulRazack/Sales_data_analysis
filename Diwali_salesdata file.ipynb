```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diwali_df=pd.read_csv('Diwali Sales Data.csv',encoding='ISO-8859-1')
diwali_df.head()
```

```python
diwali_df.shape
```

```python
diwali_df.info()
```

```python
#drop blank columns
diwali_df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#check null values
diwali_df.isna().sum()
```

```python
#drop null values
diwali_df.dropna(inplace=True)
```

```python
diwali_df['Amount']=diwali_df['Amount'].astype('int')
```

```python
diwali_df['Amount'].dtypes
```

```python
diwali_df.columns
```

```python
diwali_df.describe()
```

# Exploratory Data Analysis

# Gender

```python
#lets plot the counts of gender
ax=sns.countplot(data=diwali_df,x='Gender')
#inorder to get the counts over the bar
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#now let's check how much each gender spend in total
sales_gender=diwali_df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gender

```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

from the above its clear that female purchasing power is more than male

# Age

```python
#lets check which age group is more
ax=sns.countplot(data=diwali_df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

# Here we can see in all age groups female purchasing power is more than male

```python
#let check which age group spends more
sale_agegroup=diwali_df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sale_agegroup
```

```python
sns.barplot(x='Age Group',y='Amount',data=sale_agegroup)
```

From the above its clear that more amount spend is by 26-35 age group and least is by 0-17 age group

# State

```python
#Total no of orders from top 10 states
order_state=diwali_df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
order_state
```

```python
#Now let's plot it
# Rotate the x-axis labels to be vertical
plt.xticks(rotation=90)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State',y='Orders',data=order_state)
```

```python
#total amount /sales from top 10 states
sale_state=diwali_df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sale_state
```

```python
#Now let's plot it
# Rotate the x-axis labels to be vertical
plt.xticks(rotation=90)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State',y='Amount',data=sale_state)
```

from the above graph we can see that most of the orders are from Uttarpradesh,Maharashtra and Karnataks respectively.

# Marital Status

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(x='Marital_Status',data=diwali_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

0-'married' and1='single'. From above its clear married people have done more purchase

```python
#Now let's check amount spend based on marital  status
sale_marital=diwali_df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sale_marital
```

```python
#lets plot
plt.figure(figsize=(7,5))
sns.barplot(x='Marital_Status',y='Amount',data=sale_marital,hue='Gender')
```

Frome above graph we can see married females have more purchasing power

# Occupation

```python
#lets check the count of each occupation
plt.xticks(rotation=90)
ax=sns.countplot(x='Occupation',data=diwali_df)
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graph more purchase is done by IT sector people and least by Agriculture 

```python
#lets check which profession people spend more
sale_occupation=diwali_df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sale_occupation
```

```python
#lets plot
plt.xticks(rotation=90)

sns.barplot(x='Occupation',y='Amount',data=sale_occupation)
```

From above graph most of the buters are working in IT sector, Healthcare, Aviation 

# Product Category

```python
plt.xticks(rotation=90)
ax=sns.countplot(x='Product_Category',data=diwali_df)
for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph most people have purchased clothing & apparel and then comes food and least purchase is on hand &power tools

```python
#now lets check on which category more amount is spend and by which gender
sale_product=diwali_df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sale_product
```

```python
plt.xticks(rotation=90)
sns.barplot(x='Product_Category',y='Amount',data=sale_product,hue='Gender')
```

Eventhough countwise clothing & apparel purchase was more high amount was spend on food by female.

```python
#top 10 most sold product
diwali_df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
